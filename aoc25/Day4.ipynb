```python
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
def surrounding (x,y):
    neighbor1 = [x, y - 1]
    neighbor2 = [x, y + 1]
    neighbor3 = [x - 1, y - 1]
    neighbor4 = [x - 1 , y ]
    neighbor5 = [x -1 , y + 1]
    neighbor6 = [x + 1 , y - 1 ]
    neighbor7 = [x + 1 , y]
    neighbor8 = [x + 1 , y + 1]
    
    neighbors = [
    neighbor1,
    neighbor2,
    neighbor3, 
    neighbor4, 
    neighbor5, 
    neighbor6,
    neighbor7,
    neighbor8,
    ]
    
    return neighbors

```

```python
#checking if the function works
surrounding(2,2)
```

```python
accessible = 0

for lines in range(0, (len(diagram))):
    line = diagram[lines]
    

    for roll in range(len(line)): #line:
        position = line[roll]   #line.index(roll)
        if position == '@':
            x = lines
            y = roll
            neighbors =(surrounding(x,y))
            count = 0
            for n in neighbors:
                nx = n[0]
                ny = n[1]
                if nx < 0:
                    continue
                if nx >= len(diagram):
                    continue
                if ny < 0:
                    continue
                if ny >= len(diagram[nx]):
                    continue
                if diagram[nx][ny] == '@':
                    count = count + 1
            if count < 4:
                accessible = accessible +1
        
print(accessible, 'rolls can be taken')

```
