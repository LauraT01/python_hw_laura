```python
coordinates = []
for lines in range(0, (len(diagram))):
    line = diagram[lines]
    #print(line)

    for roll in range(len(line)): #line:
        position = roll   #line.index(roll)
        coordinate = [lines, position]
        coordinates.append(coordinate)


coordinates

```

```python
def surrounding (x,y):
    neighbor1 = [x, y - 1]
    neighbor2 = [x, y + 1]
    neighbor3 = [x - 1, y - 1]
    neighbor4 = [x - 1 , y ]
    neighbor5 = [x -1 , y + 1]
    neighbor6 = [x + 1 , y - 1 ]
    neighbor7 = [x + 1 , y]
    neighbor8 = [x + 1 , y + 1]
    
    neighbors = [
    neighbor1,
    neighbor2,
    neighbor3, 
    neighbor4, 
    neighbor5, 
    neighbor6,
    neighbor7,
    neighbor8,
    ]
    
    return neighbors

```

```python
surrounding(2,2)
```

```python
for lines in range(0, (len(diagram))):
    line = diagram[lines]
    

    for roll in range(len(line)): #line:
        position = line[roll]   #line.index(roll)
        if position == '@':
            x = lines
            y = roll
            surrounding(x,y)
            print(surrounding(x,y))
            #for adjoining in surrounding(x,y):
            print('found a paper roll in line:', lines, 'in position:', roll)
        else:
            print('no paper roll')
```
