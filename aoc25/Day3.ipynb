```python
batteries = [
'987654321111111',
'811111111111119',
'234234234234278',
'818181911112111']
```

```python
jolts = []

for battery in batteries:
    max1 = -1
    max2 = -1
    pos = -1

    for position in range(0, len(battery)):
       jolt = int((battery[position]))
       if jolt > max1:
           jolts.append(jolt)
           max1 = jolt
           pos = position

           for position2 in range(pos +1, len(battery)):
               jolt2 = int((battery[position2]))
               if jolt2 > max2:
                   jolts.append(jolt2)
                   max2 = jolt2

jolts    
```
