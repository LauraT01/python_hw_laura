```python
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
batteries =read_file_into_lines('/Users/laura/Desktop/python_hw/aoc25/input_day3')
#'987654321111111',
#'811111111111119',
#'234234234234278',
#'818181911112111'


```

```python
jolts = 0

for battery in batteries:
    max1 = max(battery[:-1])          
    pos = battery.index(max1)         

    if pos + 1 < len(battery):
        max2 = max(battery[pos + 1:])
    else:
        max2 = battery[-1]                  

    joltage = str(max1) + str(max2)
    jolts = jolts + int(joltage)

print(jolts)




                
```

### Part 2

```python
jolts = 0

for battery in batteries:
    joltage = ""
    start_index = 0

    for pick in range(12):
        remaining = battery[start_index:]
        if remaining == "":
            break

        end = len(battery) - (12 - pick) + 1
        searching = battery[start_index:end]
        max_digit = max(searching)
        relative_pos = searching.index(max_digit)
        pos = start_index + relative_pos
        joltage = joltage + max_digit
        start_index = pos + 1
    jolts = jolts + int(joltage)

print(jolts)





           

```
