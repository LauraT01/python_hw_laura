```python
IDs = ('11-22,95-115,998-1012,1188511880-1188511890,222220-222224, 1698522-1698528,446443-446449,38593856-38593862,565653-565659, 824824821-824824827,2121212118-2121212124')
```

```python
ranges = list(IDs.split(','))
```

```python
ranges
```

```python
invalidIDs = 0
for r in ranges:
    start, stop = r.split('-')
    #print(start, stop)
    for number in range(int(start), int(stop) + 1):
        #print(number)
        ID = str(number) 
        half = (len(ID) // 2)
        if ID[half:] == ID[:half]:
            invalidIDs = invalidIDs +1
        

invalidIDs
```
