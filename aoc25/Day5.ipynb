```python
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines

given = read_file_into_lines('/Users/laura/Desktop/python_hw/aoc25/input_day5')
#'3-5',
#'10-14',
#'16-20',
#'12-18',
#''
#'1',
#'5',
#'8',
#'11',
#'17',
#'32']
#read_file_into_lines('/Users/laura/Desktop/python_hw/aoc25/input_day5')

given
```
