```python
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python

rotations = read_file_into_lines('/Users/laura/Desktop/python_hw/aoc25/input_day1')
#-68, -30, 48, -5, 60, -55, -1, -99, 14, -82
#'L68', 'L30', 'R48', 'L5', 'R60', 'L55', 'L1', 'L99', 'R14', 'L82'

len(rotations)

```

```python
dial = 50
zeros = 0

for item in rotations:
    direction = item[0]
    number = int(item[1:])
    if number > 100:
        number = number % 100

    if direction == 'L':
        dial = dial - number
    else:
        dial = dial + number

    if dial > 99:
        dial = dial -100
    if dial < 0:
        dial = dial + 100
    if dial == 0: 
        zeros = zeros + 1

    

print('amount of zeros:', zeros)

    
```
