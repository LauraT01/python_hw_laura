## Rosalind Complementing a Strand of DNA
Given: A DNA string s of length at most 1000 bp.

Return: The reverse complement sc of s.

I was only able to solve this after seeing the solution for HAMM on the SW carpentry course and solving about 10 error messages, it works but it's probably not the most elegant way.

```python
s_example = "AAAACCCGGT"
reversed_example = s_example[::-1]
print(reversed_example)
sc_example = ""
for base in reversed_example:
    if base == "A": sc_example += "T"
    if base == "C": sc_example += "G"
    if base == "G": sc_example += "C"
    if base == "T": sc_example += "A"

print(sc_example)
```
